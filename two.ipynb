```python
import torch
from transformers import pipeline
import pandas as pd

t5 = "google/flan-t5-base"

df = pd.read_csv("two.csv")
```

```python
pipe = pipeline(
    "text2text-generation",
    model=t5,
    max_new_tokens=512,
    dtype= torch.float16,
)

def generate_article(topic):
    topic = "Impact of artificial intelligence on healthcare"

    prompt = f"""
    Write a well structured article of atleast 200 words on the given topic:
    "{topic}"

    Include:
    - Introduction
    - 3 to 4 key points
    - Conclusion
    """
    
    output = pipe(prompt)[0]["generated_text"]
    return output

articles = [ ]

for topic in df["topic"]:
    text = generate_article(topic)
    
    articles.append(text)

df["synthetic_article"] = articles
df.to_csv("synthetic_dataset.csv", index=False)

print("done")
```
