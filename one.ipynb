```python
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM, TextStreamer, BitsAndBytesConfig

# phi = "microsoft/Phi-3.5-mini-instruct"
t5 = "google/flan-t5-base"

```

```python
pipe = pipeline(
    "text2text-generation",
    model=t5,
    max_new_tokens = 512,
    dtype = torch.float16,
)

topic = "Impact of artificial intelligence on healthcare"

prompt = f"""
Write a well structured article of atleast 200 words on the given topic:
"{topic}"

Include:
- Introduction
- 3 to 4 key points
- Conclusion
"""

output = pipe(prompt)
text = output[0]["generated_text"]

print(text)

print("\nWord count : ", len(text.split()))
```

```python
with open("synthetic_article.txt", "w", encoding="utf-8") as f:
    f.write(text)
```
